```python
import pandas as pd
import csv
```

```python
#构建训练集
bugs = pd.read_csv(r'F:\post graduate\raw_data\data_1\buglist1.csv',encoding = 'latin-1')
Assignee = bugs['Assignee']
labels = set(Assignee)
classification = bugs['Classification']
product = bugs['Product']
component = bugs['Component']
summary = bugs['Summary']
#ID = [i for i in range(620)]
#label = dict(zip(ID,labels))
with open(r'F:\post graduate\raw_data\train.csv','a',newline='',encoding='latin-1') as csvfile:
    writer = csv.writer(csvfile)
    for i in range(len(Assignee)):
        writer.writerow([classification[i]+' '+product[i]+ ' '+component[i]+' '+summary[i],Assignee[i]])
```
