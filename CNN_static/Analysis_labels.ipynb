```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
file_path = r'F:\post graduate\dataset\data2.csv'
bug = pd.read_csv(file_path,encoding = 'latin-1')
assignee = bug['y_label']
cnt = Counter(assignee)
name = list()
fix_num = list()
for i in cnt.most_common():
    name.append(i[0])
    fix_num.append(i[1])
x = list(range(len(name)))
plt.plot(x,fix_num)
plt.xlabel('fixer')
plt.ylabel('num of fixed bugs')
plt.show()
```

```python
count = 0
total = len(assignee)
people = 0
for num in fix_num:
    count += num
    people += 1
    print(people,'%.2f' %(people/len(name)*100)+'%','fixer','fixed','%.2f' % ((count/total)*100)+'%','bugs')
```

```python
filterd_labels = list()
for i in cnt.most_common(250):
    filterd_labels.append(i[0])
```

```python
import csv
data = pd.read_csv(r'F:\post graduate\dataset\data.csv',encoding = 'latin-1')
```

```python
for i in range(len(data['x_input'])):
    if data['y_label'][i] in filterd_labels:
        with open(r'F:\post graduate\dataset\filted_data.csv','a',newline='',encoding='latin-1') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow([data['x_input'][i],data['y_label'][i]])
```
